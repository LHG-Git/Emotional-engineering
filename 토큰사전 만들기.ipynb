```python
import sentencepiece as spm
def train_sentencepiece(corpus, prefix, vocab_size=5000):
    """
    sentencepiece를 이용해 vocab 학습
    :param corpus: 학습할 말뭉치
    :param prefix: 저장할 vocab 이름
    :param vocab_size: vocab 개수
    """
    spm.SentencePieceTrainer.train(
        f"--input={corpus} --model_prefix={prefix} --vocab_size={vocab_size + 7}" +  # 7은 특수문자 개수
        " --model_type=unigram" +
        " --max_sentence_length=999999" +  # 문장 최대 길이
        " --pad_id=0 --pad_piece=[PAD]" +  # pad token 및 id 지정
        " --unk_id=1 --unk_piece=[UNK]" +  # unknown token 및 id 지정
        " --bos_id=2 --bos_piece=[BOS]" +  # begin of sequence token 및 id 지정
        " --eos_id=3 --eos_piece=[EOS]" +  # end of sequence token 및 id 지정
        " --user_defined_symbols=[SEP],[CLS],[MASK]" +  # 기타 추가 토큰 SEP: 4, CLS: 5, MASK: 6
        " --input_sentence_size=100000" +  # 말뭉치에서 셈플링해서 학습
        " --shuffle_input_sentence=true")  # 셈플링한 말뭉치 shuffle
```

```python
# morph 단위로 분할된 말뭉치 생성,  mecab으로 진행시 20분 정도 소요 
from tqdm.notebook import tqdm, trange
import zipfile

with open("corpus-mecab.txt", "w") as o_f:
    with zipfile.ZipFile(corpus) as z:
        with z.open('corpus.txt') as f:
            for i, line in enumerate(tqdm(f, total=count)):
                line = line.decode('utf-8').strip()
                tokens = mecab.morphs(line)
                string = " ".join(tokens)
                o_f.write(string)
                o_f.write("\n")
```

```python
text = """많이 서운하셨겠어요.
친구들에게 서운함을 느꼈을 때 해결하는 방법이 있나요?
카페에 가서 대화하거나 같이 술 마시면서 이야기했던 것 같아.
대화로 문제를 해결해오셨던 거군요?
그렇다면 이번 일에는 어떻게 하실 예정인가요?
직접 서운한 감정을 친구에게 얘기하려고 해.
이번에도 대화를 통해 불편한 마음을 풀려고 하시는군요.
"""
```

```python
# 말뭉치 사전 업로드
import sentencepiece as spm
sp = spm.SentencePieceProcessor()
sp.load(os.path.join(data_dir,'corpus', "corpus_5000.model"))
sp_morph = spm.SentencePieceProcessor()
sp_morph.load(os.path.join(data_dir,'corpus', "corpus_mecab_5000.model"))
```

```python
#전처리 없이 실행
tokens_A = sp.encode_as_pieces(text)
print(tokens_A)
#전처리 후 mecab 사용
tokens_B = sp_morph.encode_as_pieces(text)
print(tokens_B)
```

```python
전처리를 하지 않은 사전으로 토큰화한 결과에서는 의미를 담고 있는 단어는 붙이고 조사들은 나누는 식이었으면, mecab 전처리를 한 사전을 가지고 토큰화를 하면 조사를 나누고 의미를 담고 있는 단어도 
세밀하게 나누어서 의미 또한 나눴습니다. 이로 인해서 전처리를 하지 않은 사전으로 토큰화 했을때 보다 mecab 전처리를 한 사전을 가지고 토큰화 했을 때가 더욱 단어들을 세분화하여 나눠서 전보다 더욱 
정교화한 분석을 할 수 있었습니다.
```
